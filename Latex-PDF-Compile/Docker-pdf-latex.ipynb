```python
import docker
import time

def compile_latex_to_pdf(file_name):
    start_time = time.time()
    client = docker.from_env()

    latex_file_path = f"{file_name}.tex"
    output_pdf_path = f"{file_name}.pdf"

    # Define your LaTeX Docker image
    image = "blang/latex:ctanfull"

    # Mount the current directory to the container so it can access the .tex file and output the .pdf
    volumes = {"/path/to/your/directory": {"bind": "/data", "mode": "rw"}}

    try:
        # Run the LaTeX compiler in the Docker container
        container = client.containers.run(
            image,
            command=f"pdflatex -interaction=nonstopmode /data/{latex_file_path}",
            volumes=volumes,
            working_dir="/data",
            remove=True,
            stdout=True,
            stderr=True
        )
        output = container.decode("utf-8")
        print(output)

    except docker.errors.ContainerError as e:
        print(f"Failed to compile LaTeX file. Error: {e}")
        return False
    else:
        print(f"Compilation successful. PDF saved as {output_pdf_path}.")

        end_time = time.time()
        print("\nExecution time:", end_time - start_time)
        return True

# Note: Adjust "/path/to/your/directory" to the actual directory path where your LaTeX files are located.
```

```python
# This script is for testing a successful compilation
result = compile_latex_to_pdf("ProblemSet1")  # Assuming a valid ProblemSet1.tex exists in the specified directory

if result:
    print(f"Test Success Result: {result}")
else:
    print(f"Test Failure Result: {result}")
```
